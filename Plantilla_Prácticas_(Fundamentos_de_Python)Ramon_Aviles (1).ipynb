```
# COMIENZA AQU√ç TU C√ìDIGO #

import pandas  as pd
import sklearn as sk
from sklearn.neural_network import MLPClassifier

#  üí° --- Carga el archivo easy_telco.csv con Pandas. 

easy_telco = pd.read_csv('easy_telco.csv')

#  üí° --- Visualiza el contenido de la tabla.

display(easy_telco.head(5))

#  üí° --- Mapea las variables binarias.
#  Hay unas cu√°ntas columnas con variables binarias (Female/Male y Yes/No), utiliza la
#  funci√≥n map() de Pandas para sustituir sus valores por 0 y 1.

gender_to_num = {'Female' : 0, 'Male' : 1}
Partner_to_num = {'Yes' : 0, 'No' : 1}
Dependents_to_num = {'Yes' : 0, 'No' : 1}
PhoneService_to_num = {'Yes' : 0, 'No' : 1}
MultipleLines_to_num =  {'Yes' : 0, 'No' : 1}
OnlineSecurity_to_num =  {'Yes' : 0, 'No' : 1}
OnlineBackup_to_num =  {'Yes' : 0, 'No' : 1}
DeviceProtection_to_num = {'Yes' : 0, 'No' : 1}
TechSupport_to_num = {'Yes' : 0, 'No' : 1}
StreamingTV_to_num = {'Yes' : 0, 'No' : 1}
StreamingMovies_to_num = {'Yes' : 0, 'No' : 1}
PaperlessBilling_to_num = {'Yes' : 0, 'No' : 1}
Churn_to_num = {'Yes' : 0, 'No' : 1}

easy_telco['gender'] = easy_telco['gender'].map(gender_to_num)
easy_telco['Partner'] = easy_telco['Partner'].map(Partner_to_num)
easy_telco['Dependents'] = easy_telco['Dependents'].map(Dependents_to_num)
easy_telco['PhoneService'] = easy_telco['PhoneService'].map(PhoneService_to_num)
easy_telco['MultipleLines'] = easy_telco['MultipleLines'].map(MultipleLines_to_num)
easy_telco['OnlineSecurity'] = easy_telco['OnlineSecurity'].map(OnlineSecurity_to_num)
easy_telco['OnlineBackup'] = easy_telco['OnlineBackup'].map(OnlineBackup_to_num)
easy_telco['DeviceProtection'] = easy_telco['DeviceProtection'].map(DeviceProtection_to_num)
easy_telco['TechSupport'] = easy_telco['TechSupport'].map(TechSupport_to_num)
easy_telco['StreamingTV'] = easy_telco['StreamingTV'].map(StreamingTV_to_num)
easy_telco['StreamingMovies'] = easy_telco['StreamingMovies'].map(StreamingMovies_to_num)
easy_telco['PaperlessBilling'] = easy_telco['PaperlessBilling'].map(PaperlessBilling_to_num)
easy_telco['Churn'] = easy_telco['Churn'].map(Churn_to_num)


#  üí° --- Visualiza el contenido de la tabla.

display(easy_telco.head(5))

#  üí° --- Convierte las variables no binarias a binarias.
#  Las columnas InternetService, Contract y PaymentMethod contienen m√°s
#  de una categor√≠a distinta y por tanto no podemos mapearlas directamente a 0 y 1.
#  La estrategia a seguir ser√° usar el m√©todo pd.get_dummies(...) que nos convertir√°
#  cada valor de dicha columna en columnas binarias. Aqu√≠ tienes el ejemplo de c√≥mo
#  transformar una de estas columnas, comprende el c√≥digo y realiza lo mismo con las 
#  restantes.

dumm  = pd.get_dummies(easy_telco['InternetService'], prefix='internet_')
easy_telco = pd.concat([easy_telco, dumm], axis=1)
easy_telco = easy_telco.drop('InternetService', axis=1)

dumm  = pd.get_dummies(easy_telco['Contract'], prefix='Contract_')
easy_telco = pd.concat([easy_telco, dumm], axis=1)
easy_telco = easy_telco.drop('Contract', axis=1)

dumm  = pd.get_dummies(easy_telco['PaymentMethod'], prefix='PaymentMethod_')
easy_telco = pd.concat([easy_telco, dumm], axis=1)
easy_telco = easy_telco.drop('PaymentMethod', axis=1)

# ...

#  üí° --- Visualiza el contenido de la tabla.

display(easy_telco.head(5))
#  üí° --- Elimina la columna customerID pues no aporta informaci√≥n relevante.

neweasy_telco = easy_telco.drop(columns="customerID")

#  üí° --- Convierte tu tabla de datos a matrices Numpy.

numpyeasy_telco = neweasy_telco.values

print(numpyeasy_telco)

#  Create una matriz de datos de entrada, que contenga todas las columnas menos 'Churn'

neweasy_telco_All = neweasy_telco.drop(columns = 'Churn')


#  y una matriz de salida que s√≥lo contenga a 'Churn'.

neweasy_telco_Churn = neweasy_telco.drop(columns = ['gender',
                                                    'SeniorCitizen','Partner','Dependents','tenure',
                                                    'PhoneService','MultipleLines','OnlineSecurity','OnlineBackup',
                                                    'DeviceProtection','TechSupport','StreamingTV','StreamingMovies',
                                                    'PaperlessBilling','MonthlyCharges','TotalCharges','internet__DSL'
                                                    ,'internet__Fiber optic','internet__No','Contract__Month-to-month',
                                                    'Contract__One year','Contract__Two year',
                                                    'PaymentMethod__Bank transfer (automatic)',
                                                    'PaymentMethod__Credit card (automatic)',
                                                    'PaymentMethod__Electronic check',
                                                    'PaymentMethod__Mailed check'])


display(neweasy_telco_All)
display(neweasy_telco_Churn)
#  üí° --- Utiliza Sklearn para crear un modelo, entrenarlo y evaluarlo. 

import sklearn as sk
from sklearn.linear_model import LogisticRegression

easy_telco_test = neweasy_telco_All.values
easy_telco_test1 = neweasy_telco_Churn.values

X = easy_telco_test
Y = easy_telco_test1

print(X.shape, Y.shape)

clf = LogisticRegression()

clf.fit(X, Y)

# 3¬∫. Con el modelo ajustado, puedes hacer predicciones.

yp = clf.predict(X)
print(yp, '\n')

# 4¬∫ Con el modelo ajustado, puedes evaluar su rendimiento.

ev = clf.score(X, Y)
print(ev, '\n')

#  Deberemos de obtener una tasa de acierto superior al 70%.

from sklearn.neural_network import MLPClassifier

red_neuronal = MLPClassifier(hidden_layer_sizes=(50,), activation='relu', max_iter=2000)  # -- Red neuronal ya creada.

red_neuronal.fit(X, Y)

rnp = red_neuronal.predict(X)

rnev = red_neuronal.score(X,Y)

print (rnp,rnev)

#  (Opcional: Para considerar v√°lido la evaluaci√≥n del modelo, realmente tendr√≠amos
#  que valorar si nuestra modelo est√° generalizando lo aprendido. Para ello es una
#  buena pr√°ctica crearte una dataset de entrenamiento y otro de prueba. Si quieres
#  puedes probar a generarlos con la funci√≥n train_test_split y volver a entrenar
#  y a evaluar a tu modelo.)


from sklearn.model_selection import train_test_split

easy_telco_test = neweasy_telco_All.values
easy_telco_test1 = neweasy_telco_Churn.values

X = easy_telco_test
Y = easy_telco_test1

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.33)

print(X_train.shape, X_test.shape, Y_train.shape, Y_test.shape)
clf.fit(X_train, Y_train)
yp = clf.predict(X_test)
ev = clf.score(X_test, Y_test)

red_neuronal = MLPClassifier(hidden_layer_sizes=(50,), activation='relu', max_iter=2000)  # -- Red neuronal ya creada.

red_neuronal.fit(X_train, Y_train)

rnp = red_neuronal.predict(X_test)
rnev = red_neuronal.score(X_test,Y_test)

print (rnp,rnev)
```
