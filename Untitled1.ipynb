# **Buscando la predicci√≥n del producto mas probable de tener una ganancia **

```python
#creando el dataframe tesis1

tesis1 = pd.read_csv('/content/drive/MyDrive/Tesis3.csv')
tesis1.head()
```

```python
#Creando la variable Ganancia_alta
tesis1['Ganancia_alta'] = np.where(tesis1.Ganancia > average(tesis1.Ganancia), 0, 1)
# Una vez creada la nueva variable se descarta la original
tesis1 = tesis1.drop(columns = 'Ganancia')
```

```python
# Divisi√≥n de los datos en train (entrenamiento) y test (prueba)

X_train, X_test, y_train, y_test = train_test_split(
                                        tesis1.drop(columns = ['Ganancia_alta','Nombre_cliente','Producto_Id','PrecioUnitario','TotalCostoVenta','Cantidad',
                                                              'FechaEmision','Clasificacion_producto','TotalPrecio',
                                                              'NombreCatClient','Nombrecat01Clien','Nombre_empleado','Empleado_Id','Cod_empleado',
                                                               'FechaEmision','Ubigeo','UnidadNegocio_Id','Fabricante_Id',
                                                               'Marca_Id','factorventa','Tipo_cliente','Serie','Numero','CondicionComercial',
                                                               'Fabricante','Proveedor','Proveedor_Id','FuerzaVenta','grupoventa','Marca', 'Direccion_Id', 'Cliente_Id',
                                                               'Venta_Id','CategoriaCliente_Id','Territorio_Id','TipoCliente_Id','Categoria01Cliente_Id'
                                                               ]),
                                        tesis1['Ganancia_alta'],
                                        random_state = 123
                                    )

# One-hot-encoding de las variables categ√≥ricas
cat_cols = X_train.select_dtypes(include=['object', 'category']).columns.to_list()
numeric_cols = X_train.select_dtypes(include=['float64', 'int']).columns.to_list()

preprocessor = ColumnTransformer(transformers= [('onehot', OneHotEncoder(handle_unknown='ignore'), cat_cols)],
                    remainder='passthrough', verbose_feature_names_out=True
               )

#con el m√©todo fit() se aprenden las transformaciones con los datos de entrenamiento y se aplican a
# los dos conjuntos con transform(). Ambas operaciones a la vez con fit_transform().
X_train_prep = preprocessor.fit_transform(X_train)
X_test_prep  = preprocessor.transform(X_test)
```

```python
# Convertir el output del ColumnTransformer en dataframe y a√±adir el nombre de las columnas

# Nombre de todas las columnas
encoded_num = preprocessor.named_transformers_['onehot'].get_feature_names_out(cat_cols)
labels = np.concatenate([numeric_cols, encoded_num])

# Conversi√≥n a dataframe
X_train_prep = pd.DataFrame(X_train_prep.toarray() , columns=labels)
X_test_prep  = pd.DataFrame(X_test_prep.toarray() , columns=labels)
X_train_prep.columns = X_train_prep.columns.str.replace('Nombre_Producto_', '')
X_test_prep.columns = X_test_prep.columns.str.replace('Nombre_Producto_', '')
X_train_prep.info()
```

```python
# Parametros del Grid 

param_grid = {'n_estimators'  : [1,3,5,20],
              'max_features'  : ['auto', 'sqrt', 'log2'],
              'max_depth'     : [1,2,5],
              'subsample'     : [1],
              'learning_rate' : [0.001,0.01,0.1]
             }

# evaluacion por grid search con validaci√≥n cruzada

grid = GridSearchCV(
        estimator  = GradientBoostingClassifier(random_state=123),
        param_grid = param_grid,
        scoring    = 'accuracy',
        n_jobs     = multiprocessing.cpu_count() - 1,
        cv         = RepeatedKFold(n_splits=3, n_repeats=1, random_state=123), 
        refit      = True,
        verbose    = 0,
        return_train_score = True
       )

grid.fit(X = X_train_prep, y = y_train)

# Resultados

resultados = pd.DataFrame(grid.cv_results_)
resultados.filter(regex = '(param*|mean_t|std_t)') \
    .drop(columns = 'params') \
    .sort_values('mean_test_score', ascending = False) \
    .head(4)
```

```python
#Hiperparametros encontrados 

print("Mejores hiperpar√°metros encontrados")
print("----------------------------------------")
print(grid.best_params_, ":", grid.best_score_, grid.scoring)
```

```python
modelo_final = grid.best_estimator_
```

```python
# Error de test del modelo final

predicciones = modelo_final.predict(X = X_test_prep)
predicciones[:20]
```

```python
mat_confusion = confusion_matrix(
                    y_true    = y_test,
                    y_pred    = predicciones
                )

accuracy = accuracy_score(
            y_true    = y_test,
            y_pred    = predicciones,
            normalize = True
           )

print("Matriz de confusi√≥n")
print(mat_confusion)
print("")
print(f"El porcentaje de acierto del test es: {100 * accuracy} %")
```

```python
# Predicci√≥n de probabilidades

predicciones = modelo_final.predict_proba(X = X_test_prep)
predicciones[:15, :]
```

```python
# Clasificaci√≥n empleando la clase de mayor probabilidad

df_predicciones = pd.DataFrame(data=predicciones, columns=['0', '1'])
df_predicciones['clasificacion_default_0.5'] = np.where(df_predicciones['0'] > df_predicciones['1'], 0, 1)
df_predicciones.head()
```

```python
# Clasificaci√≥n final empleando un threshold de 0.7 para la clase 1.

df_predicciones['clasificacion_custom_0.7'] = np.where(df_predicciones['1'] > 0.8, 1, 0)
df_predicciones[df_predicciones['clasificacion_custom_0.7']!=df_predicciones['clasificacion_default_0.5']] \
    .head()
```

```python
#Listado de producto con su importancia calculada en el modelo 
importancia_predictores = pd.DataFrame(
                            {'Producto': X_train_prep.columns,
                             'importancia': modelo_final.feature_importances_}
                            )
print("Importancia de los predictores en el modelo")
print("-------------------------------------------")
importancia_predictores.sort_values('importancia', ascending=False)
```

```python
#parametros de importancia para calculo de media y desviaci√≥n
importancia = permutation_importance(
                estimator    = modelo_final,
                X            = X_train_prep,
                y            = y_train,
                n_repeats    = 5,
                scoring      = 'neg_root_mean_squared_error',
                n_jobs       = multiprocessing.cpu_count() - 1,
                random_state = 123
             )

# Se almacenan los resultados (media y desviaci√≥n) en un dataframe llamada df_importancia
df_importancia = pd.DataFrame(
                    {k: importancia[k] for k in ['importances_mean', 'importances_std']}
                 )
df_importancia['feature'] = X_train_prep.columns
df_importancia.sort_values('importances_mean', ascending=False)
```

```python
# Creacion del Gr√°fico
fig, ax = plt.subplots(figsize=(10, 30))
df_importancia = df_importancia.sort_values('importances_mean', ascending=True)
ax.barh(
    df_importancia['feature'],
    df_importancia['importances_mean'],
    xerr=df_importancia['importances_std'],
    align='center',
    alpha=0
)
ax.plot(
    df_importancia['importances_mean'],
    df_importancia['feature'],
    marker="D",
    linestyle="",
    alpha=0.8,
    color="r"
)
ax.set_title('Importancia de los productos')
ax.set_xlabel('Incremento del error');
```
